```python
def ganti_isi_column(merged_wisata,merged_wisata_review):
  for i in range(len(merged_wisata)) :
    merged_wisata[i] = merged_wisata_review[i]
  print(merged_wisata)
```

```python
def preprocessing(document_to_be_scored):
  pisah = {}
  for i in range(len(document_to_be_scored)):
    pisah[i] = document_to_be_scored[i].split() #melakukan split

  #menghilangkan stopword memakai library nltk
  stopword=['wisata']#list stopword

  removed = {}
  hasil = []
  listStopword = stopwords.words('indonesian') #list stopword

  #tambahkan stopword
  for i in stopword:
    listStopword.append(i)
  # print(listStopword)

  # listStopword = set(stopwords.words('indonesian')) #list stopword
  hitung = 0 #untuk menghitung frekuensi
  for i in pisah.values():
    for j in i:
      if j not in listStopword:#jika setiap kata tidak ada di list stopword
        hasil.append(j)
    removed[hitung] = hasil[:]
    del hasil[:]
    hitung=hitung+1

  #memakai library sastarawi
  #memakai library sastrawi untuk melakukan stemming
  factory = StemmerFactory() #membuat objek stemmer
  stemmer = factory.create_stemmer()
  hitung = 0

  hasil=[]
  hasil2 = {}

  for i in removed.values():
    for j in i:
      hasil.append(stemmer.stem(j))#hasil stemming
    hasil2[hitung] = hasil[:]
    del hasil[:]
    hitung = hitung+1
  hasil =""
  korpus2 =[]
  for i in hasil2.values():
    hasil =""
    for j in i:
      hasil = hasil+ " " +j
    korpus2.append(hasil)
  ganti_isi_column(document_to_be_scored,korpus2)#timpa hasil

```

```python
# refrensi
# https://towardsdatascience.com/comparing-documents-with-similarity-metrics-e486bc678a7d
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# returns the cosine similarity value of the two given texts (mengembalikan nilai cosine similarity dari dua teks yang diberikan)
def compute_cosine_similarity(text1, text2):

    # stores text in a list
    list_text = [text1, text2]

    # converts text into vectors with the TF-IDF
    vectorizer = TfidfVectorizer()
    vectorizer.fit_transform(list_text)
    tfidf_text1, tfidf_text2 = vectorizer.transform([list_text[0]]), vectorizer.transform([list_text[1]])

    # computes the cosine similarity
    cs_score = cosine_similarity(tfidf_text1, tfidf_text2)

    return np.round(cs_score[0][0],2)
```

```python
#masukkan input
query =''
query = input('masukkan query: ')
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
factory = StemmerFactory() #membuat objek stemmer
stemmer = factory.create_stemmer()
query =stemmer.stem(query)
```

```python
similarity = {}

i =0
for doc in merged_wisata_prep['Review']: #similarities antara input dan review
  similarity[i]= compute_cosine_similarity(query,doc)
  i = i+1
  # doc_id.append(str(i))
output = sorted(similarity.items(), key=lambda x:x[1], reverse=False)
print(output)
print("menampilkan 10 Data paling relevan (nilai cosine similiarity tertinggi) dengan "+ "\'" +query+"\'")
i =0
for k, v in output:
  # if v ==0 :continue
  if v < 0.05 :continue #jika nilai scoring dibawah 0.05 skip
  if(i > 10): break
  try:
    print("pada record ke ",str(k))
    print("dengan nilai similaritas: ", v)
    print("berikut adalah descriptionnya: ", merged_wisata['description'][k])
    print("berikut adalah reviewnya: " , merged_wisata['Review'][k])
    print()
  except:
    continue
  i=i+1
```
